```python
from transformers import pipeline, set_seed

generator = pipeline('text-generation', model='gpt2')
set_seed(42)

print("simple example")
prompt = "AI can"
results = generator(prompt, max_length = 50 , num_return_sequences=1)
print(results[0]['generated_text'])
```

```python
print("\n 互动文本生成")
def interactive_advanced_generator():
  while True:
    prompt = input("\n请输入一个文本开头（输入'quit'退出）")
    if prompt.lower() == 'quit':
      break

    length = input("请输入生成的最大长度（默认100）")
    length = int(length) if length.isdigit() else 100

    num_samples = input("请输入生成样本数（1-3 默认为1）")
    num_samples = int(num_samples) if num_samples.isdigit() and 1 <= int(num_samples) <=3 else 1

    print("\n 生成中")
    results = generator(
        prompt,
        max_length = length,
        num_return_sequences = num_samples,
        temperature = 0.7,
        do_sample = True
    )

    for i, result in enumerate(results):
      print(f"\n样本 {i+1}:")
      print(result['generated_text'])

interactive_advanced_generator()


```

```python
print("\n尝试不同的模型")
print("让我们试试一个专门用于回答的视频")

question_answerer = pipeline("question-answering", model = "distilbert-base-cased-distilled-squad")

def interactive_qa():
  context = """为什么天空是蓝色的？
我们每天抬头看到的天空，大多数时候是蓝色的。其实，这和阳光的组成以及大气的性质有关。阳光并不是单一颜色，而是由许多不同颜色的光混合而成。这些光的波长从短到长依次包括紫、蓝、绿、黄、橙、红等。

当阳光进入地球大气层时，会遇到空气分子和微小的颗粒。波长较短的光（比如紫光和蓝光）更容易被这些分子散射，而波长较长的光（比如红光）则不容易散射太多。虽然紫光的波长比蓝光更短，但人眼对蓝色更敏感，且大气对紫光的吸收较多，因此我们感受到的天空主要是蓝色。

在日出和日落时，阳光需要穿过更厚的大气层，短波光被大量散射掉，剩下的长波红光和橙光进入眼睛，所以此时的天空会呈现出迷人的暖色调。"""

  print("\n已知信息：")
  print(context)

  while True:

    question = input("\n请输入一个问题（输入'quit'退出）：")
    if question.lower() == 'quit':
      break

    result = question_answerer(question = question, context = context)
    print(f"{result['answer']}")
    print(f"{result['score']:.4f}")


interactive_qa()
```
